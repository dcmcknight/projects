# List and Dictionaries

```python
my_dict = {}
my_dict['Antonio'] = 26
my_dict['Brandon'] = 25
my_dict['Sterling'] = 26
my_dict['Jarvis'] = 24
my_dict['Mariela'] = 23
my_dict
```

```python
data = {
    'Antonio': [89, 75, 93, 98],
    'Brandon': [45, 67, 88, 91],
    'Sterling': [55, 99, 100, 65],
    'Jarvis': [77, 88, 87, 90],
    'Mariela': [90,88, 65, 45]
         }
print(sum(data['Antonio'])/len(data['Antonio']))
print(sum(data['Brandon'])/len(data['Brandon']))
print(sum(data['Sterling'])/len(data['Sterling']))
print(sum(data['Jarvis'])/len(data['Jarvis']))
print(sum(data['Mariela'])/len(data['Mariela']))
```

```python
def calcAverage(scores):
    name = input('Enter name: ')
    average = sum(scores[name])/len(scores[name])
    return '{:s}\'s final score is {:.2f}%'.format(name, average)

print(calcAverage(data))
```

## Reflection:

* I first create a dictionary with a list of scores associated to a particular student named 'data'
* I define a function that accepts a dictionary as a parameter named 'scores'
* created a variable called 'name' which represents the key in the dictionary
* I calculate the average by taking the sum of the scores (i.e. value of scores dict) for a specific student (i.e. key of scores dict) and divide it by the length of the scores list (i.e. value of scores dict).
* The function then returns the inputted name and average grade for that student
